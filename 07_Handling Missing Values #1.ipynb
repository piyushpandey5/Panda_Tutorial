# Handling missing values in python

```python
import pandas as pd
```

```python
df=pd.read_csv('E:\\new1\\Book3.csv')
df
```

```python
df=pd.read_csv('E:\\new1\\Book3.csv', na_values='not available')
df
```

```python
df=pd.read_csv('E:\\new1\\Book3.csv', na_values=['not available','no values'])
df
```

```python
df=pd.read_csv('E:\\new1\\Book3.csv', na_values='not available')
df
```

```python
df=pd.read_csv('E:\\new1\\Book3.csv',na_values={'Companies':'no values','Growth':'not available'})
df
```

```python
df=pd.read_csv('E:\\new1\\Book3.csv')
df
```

```python
df=pd.read_csv('E:\\new1\\Book3.csv',keep_default_na=False)
df
```

```python
df=pd.read_csv('E:\\new1\\Book3.csv',keep_default_na=True)
df
```

```python
df=pd.read_csv('E:\\new1\\Book3.csv',na_filter=False)
df
```
