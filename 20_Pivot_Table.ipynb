# Pandas Pivot Table

```python
import pandas as pd
```

```python
df=pd.read_csv('E:\\new1\\Movies.csv')
df
```

```python
df.pivot_table(index='Genre')#show the aggregate value
```

```python
df.pivot_table(index='Genre',columns='year')
```

```python
df.pivot_table(index='Genre',columns='year',aggfunc='count')
```

```python
df.pivot_table(index='Genre',columns='year',aggfunc='max')
```

```python
df.pivot_table(index='Genre',columns='year',aggfunc='sum')
```

```python
df.pivot_table(index='Genre',columns='year',fill_value='none')
```

```python
df.pivot_table(index='Genre',columns='year',fill_value='none',margins=True)
```

```python
df.pivot_table(index='Genre',columns='year',fill_value='none',margins=True,aggfunc='sum')
```
